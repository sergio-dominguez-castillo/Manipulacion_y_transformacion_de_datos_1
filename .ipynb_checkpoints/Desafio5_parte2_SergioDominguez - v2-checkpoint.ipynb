```python
# importacion de libreria
import pandas as pd
import numpy as np
```

a. Carga los datos y crea un DataFrame con ellos.

```python
df = pd.read_csv("Cleaned_DS_Jobs.csv")
df
#df.shape
```

b. Utiliza la siguiente lista de valores que serán considerados como nulos:

```python
nulos_dict={"na":np.nan, "NA":np.nan, -1:np.nan, "0":np.nan, "-1":np.nan, "null":np.nan, "n/a":np.nan, "N/A":np.nan, "NULL":np.nan}
df_replaced = df.replace(nulos_dict)
df_replaced
```

c. Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna())

```python
# reviso los nan por columnas
#df_replaced.isna().sum()
# elimino registros con nan
df_replaced=df_replaced.dropna(axis = 0)
```

d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo
y Máximo. (hint: Utiliza el método apply sobre la columna.)

```python
#df_replaced['Salary Estimate'].str.split(pat='-',expand=True)
#df_replaced[['Salario Estimado Minimo','Salario Estimado Maximo']] = 
df_replaced['Salario Estimado Minimo']=df_replaced['Salary Estimate'].apply(lambda x: x.split("-")).str[0]
df_replaced['Salario Estimado Maximo']=df_replaced['Salary Estimate'].apply(lambda x: x.split("-")).str[1]
print(df_replaced)
# df_replaced['Salario Estimado'].apply()
#df_1 = df_replaced['Salario Estimado'].apply(lambda x: x.split('-')
                                             
                                             #, axis=1, result_type="expand")
```

```python
# genero un nuevo df con las columnas separadas del salario estimado
df_minmax=df_replaced['Salary Estimate'].str.split('-', expand=True)
# renombro las columnas
df_minmax=df_minmax.rename(columns={0:'Salario Estimado Minimo',1:'Salario Estimado Maximo'})
# reviso el tipo de dato
print(df_minmax.dtypes)
# tranformo a numerico los valores de las columnas
df_minmax['Salario Estimado Minimo']=pd.to_numeric(df_minmax['Salario Estimado Minimo'])
df_minmax['Salario Estimado Maximo']=pd.to_numeric(df_minmax['Salario Estimado Maximo'])
# reviso el tipo de dato
print(df_minmax.dtypes)
# concateno los dos df
df_final=pd.concat([df_replaced,df_minmax], axis=1)
df_final

# no logre aplicar la separacion de columna salario estimado con el metodo apply
#df_replaced[['Salario Estimado Minimo','Salario Estimado Maximo']] = df_replaced['Salario Estimado'].apply(lambda x: x.split('-') , 
#axis=1, result_type="expand")
```

e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint:
utilice reemplazo con diccionario usando el método replace sobre la columna.)

```python
# reviso el contenido de la columan size
print(df_final['Size'].value_counts())
# creo el diccionario, se añade faltante
size_dict={'10000+ employees':'Mega Empresas', '5001 to 10000 employees': 'Grandes Empresas', '1001 to 5000 employees':'Medianas Empresas',
           '201 to 500 employees':'Pequeñas Empresas', '51 to 200 employees':'Pequeñas Grandes Empresas',
           '501 to 1000 employees':'Microempresas', 'Unknown':'Empresas sin Información', '1 to 50 employees':'Pequeñas Medianas Empresa'}
df_final = df_final.replace(size_dict)
#df_replaced
print(df_final['Size'].value_counts())
```

f. Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y
la media del salario estimado máximo por tamaño de empresa. (hint: utiliza
pd.pivot_table para generar la vista adecuada con las columnas generadas.)

```python
df_pivot_reporte1 = pd.pivot_table(
                                   data = df_final # DataFrame.
                                 , index = ["Size"] # Columna(s) del DataFrame original que queremos como indice de la tabla resultante.
                                 , values = ["Salario Estimado Minimo", "Salario Estimado Maximo"] # valores a lo que le aplicaremos la aggfunc.
                                 , aggfunc= "mean" # función a aplicar a la columna de valores.
                                 , fill_value = 0  # tranforma Nan a cero
                                 , margins = True # Generar una fila/columna de totales.
)
df_pivot_reporte1
```
