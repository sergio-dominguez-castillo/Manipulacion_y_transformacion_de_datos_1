```python
df_officers = pd.read_pickle("officers.pkl")
#print(df_officers.head())
print(df_officers.shape)
```

b. Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las
columnas que se llaman igual usando el parámetro suffixes de pd.merge().

```python
df_staging = df_incidents.merge(df_subjects, on='case_number', how='left', suffixes=('', '_sub'))
# print(df_staging.head())  
print(df_staging.shape)
df_staging.head()
```

```python
df_final = df_staging.merge(df_officers, on='case_number', how='left', suffixes=('_sub', '_off'))
#print(df_final.head())  
print(df_final.shape)
df_final.head()
```

f. Genera una tabla pivote que muestre el número de casos por mes en la fila y por año en
las columnas. hint: genere columnas de tipo fecha con pd.to_datetime(). ¿Se ve alguna
relación entre mes del año y cantidad de casos?

```python
df_final.dtypes
# la columna date no esta en formato para trabajar por fecha
```

```python
# se añade columna date2 en formato para trabajar con fecha
df_final['date2']= pd.to_datetime(df_final['date'])
#df_final.dtypes
```

```python
# ahora añado columnas de mes y año
df_final['anno']= df_final['date2'].dt.year
df_final['mes']= df_final['date2'].dt.month
#df_final.dtypes
```

```python
df_pivot_reporte1 = pd.pivot_table(
                                   data = df_final # DataFrame.
                                 , index = ["mes"] # Columna(s) del DataFrame original que queremos como indice de la tabla resultante.
                                 , columns = "anno" # columnas
                                 , values = "case_number" # valores a lo que le aplicaremos la aggfunc.
                                 , aggfunc= "count" # función a aplicar a la columna de valores.
                                 , fill_value = 0  # tranforma Nan a cero
                                 , margins = True # Generar una fila/columna de totales.
)
df_pivot_reporte1
```

¿Se ve alguna relación entre mes del año y cantidad de casos?
a primera vista no veo relacion, lo si veo son casos puntuales muy altos respecto de otros meses dic/2009 con 15 casos jun/2015 con 18 casos y jul/2016 con 14 casos

g. Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el
género del subject. ¿Cómo interpretas los valores que muestra esta vista?

```python
df_pivot_reporte2 = pd.pivot_table(
                                   data = df_final # DataFrame.
                                 , index = ["gender_off"] # Columna(s) del DataFrame original que queremos como indice de la tabla resultante.
                                 , columns = "gender_sub"
                                 , values = "case_number" # valores a lo que le aplicaremos la aggfunc.
                                 , aggfunc= "count" # función a aplicar a la columna de valores.
                                 , margins = True # Generar una fila/columna de totales.
)

df_pivot_reporte2
```

Explicacion: la tabla pivote nos muestra la cantidad de numeros de casos de los oficiales por genero que estan acumulados por el genero del sujeto, de la informacion obtenida señala que las oficiles de genero femenino han trabajado en 20 casos de los cuales 2 casos son sobre sujetos de genero femenino y 18 casos sobre sujetos de genero masculino.
